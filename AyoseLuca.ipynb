```python
import numpy as np
import pandas as pd
```

```python
df_dwells   = pd.read_csv('hack4good_dwells.csv',sep = ',',low_memory=False)
df_journeys = pd.read_csv('hack4good_journeys.csv',sep = ',',low_memory=False)
df_antenas  = pd.read_csv('hack4good_antenas.csv',sep = ',',low_memory=False)
```

```python
df_antenas.dropna(axis=0,how='any',inplace=True)
df_dwells.dropna(axis=0,how='any',inplace=True)
df_journeys.dropna(axis=0,how='any',inplace=True)

df_dwells = df_dwells[['cod_mpio','dwells','people']]
```

```python
 
#Suma de Trafico total por municipio
municipios = df_antenas['cod_mpio'].unique()
trafico = ['3g_traf_dl', '3g_traf_ul' ,'3g_completadas','3g_no_completadas' ,'2g_traf_dl', '2g_traf_ul', '2g_completadas','2g_no_completadas']

n_columns = ['cod_mpio','3g_traf_dl', '3g_traf_ul' ,'3g_completadas','3g_no_completadas' ,'2g_traf_dl', '2g_traf_ul', '2g_completadas','2g_no_completadas']

new_df = pd.DataFrame(columns=n_columns)  
for i in municipios:
    aux = df_antenas.loc[df_antenas['cod_mpio'] == i]
    
    l = [i,aux['3g_traf_dl'].mean(),aux['3g_traf_ul'].mean(),aux['3g_completadas'].mean(),aux['3g_no_completadas'].mean(),aux['2g_traf_dl'].mean(),aux['2g_traf_ul'].mean(),aux['2g_completadas'].mean(),aux['2g_no_completadas'].mean()]
   
    new_df.loc[len(new_df)] = l
    
    
new_df  
```

```python
 
municipios = df_dwells['cod_mpio'].unique()  
dwells =['cod_mpio','dwells' ,'people']
 
new_df3 = pd.DataFrame(columns=dwells) 
for i in municipios:
    aux = df_dwells.loc[df_dwells['cod_mpio'] == i]
    
    new_df3.loc[len(new_df3)] = [i,aux['dwells'].mean(),int(aux['people'].mean())]
new_df3
```

```python
merged_df = pd.merge(new_df,new_df3,how='left', on=['cod_mpio'])
merged_df
```
